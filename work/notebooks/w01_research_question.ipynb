## 1. My lane (or freestyle) and why

I'm going with the CTR / Engagement Opportunity Scoring lane. In Week 1, I already dug into how click-through rate varies across content type and position tier, and this lane felt like a natural next step — turning that exploration into something more structured: a ranked score that actually points to where the opportunity is.


```python
import os, sys, subprocess

IN_COLAB = "google.colab" in sys.modules
REPO_URL = "https://github.com/suhanisingh5617-ai/flyrank-ml-internship"
REPO_DIR = "flyrank-ml-internship"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)
    os.chdir(REPO_DIR)

import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

df['position_tier'] = pd.cut(df['avg_position'], bins=[0,3,10,20,100], labels=['top3','4-10','11-20','20+'])
result = df.groupby(['position_tier','content_type'])['ctr'].mean().sort_values()
print(result)
```

## 2. The question: decision, action, cost of a wrong call

The question I want to answer is: which pages are underperforming on CTR relative to where they rank, and how much room is there to improve? The decision this feeds into is which pages a content or SEO team should prioritize for optimization. The action would be flagging those pages for a title/meta rewrite or content refresh. If the score gets it wrong, the cost isn't huge but it's real — the team could spend time optimizing a page that was already fine, or worse, miss a high-traffic page that's genuinely underperforming.

```python
underperforming = df[df['ctr'] < df.groupby('position_tier')['ctr'].transform('median')]
print("Pages underperforming vs their tier median:", len(underperforming))
```

## 3. Quick look at the data (2-3 real numbers)

From my Week 1 discovery, I already have some evidence worth citing: in the top3 position tier, "feedly article" content had a CTR of 9.87, while "comparison article" content in the same tier had a CTR close to 0.00. That's a huge gap for pages ranking in the same spot, which suggests content type alone is doing a lot of the work here — worth digging into further.

```python
top3_feedly = df[(df['position_tier']=='top3') & (df['content_type']=='feedly article')]['ctr'].mean()
top3_comparison = df[(df['position_tier']=='top3') & (df['content_type']=='comparison article')]['ctr'].mean()

print("Top3 tier — feedly article CTR:", round(top3_feedly, 3))
print("Top3 tier — comparison article CTR:", round(top3_comparison, 3))
print("Gap:", round(top3_feedly - top3_comparison, 3))
```

## 4. Careful words: what I can and can't claim

I can say that CTR differences across content type and position are observed and directional in this dataset. I can't say I've proven why users click less on certain content types, or that fixing this will definitely raise CTR — that would need testing, not just observation.
