## 5. Why ML beats a fixed rule here

A fixed rule like "flag any page with CTR below 0.1" only looks at one signal in isolation and ignores context — a 0.1 CTR at position 1 is a real problem, but the same CTR at position 15 might be completely normal. A model can combine multiple signals (position, impressions, content type, staleness) at once and learn how they interact, catching underperformance patterns a single-threshold rule would either over-flag or completely miss.
